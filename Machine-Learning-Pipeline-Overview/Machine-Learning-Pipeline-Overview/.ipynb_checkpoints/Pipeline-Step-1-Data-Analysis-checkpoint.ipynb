# Missing Values

```python
#missing values
missing_variables = [var for var in data.columns if data[var].isnull().sum() > 0]
#determine percentage of missing values
data[missing_variables].isnull().sum()/data.shape[0]
```

```python
def plot_missing_values_target(df, missing_column, target_column, aggregation='median'):
    df = df.copy()
    
    #adding one to where value is missing
    
    missing_percentage = df[missing_column].isnull().sum()/df.shape[0]
    
    df[missing_column] = np.where(df[missing_column].isnull(),1,0)
    df.groupby(missing_column).agg({target_column:aggregation}).plot.bar()
    
    plt.title(missing_column + " {}".format(np.round(missing_percentage,2)))
    plt.show()

for var in missing_variables:
    plot_missing_values_target(data, var, "SalePrice")
    
```

```python
def analysis_year_vars(df, var, target_var="SalePrice"):
    df = df.copy()
    df[var] = df['YrSold'] - df[var]
    plt.scatter(df[var],df[target_var])
    plt.title("Relationship of {} to {}".format(var, target_var))
    plt.xlabel(var)
    plt.ylabel(target_var)
    plt.show()
    
for var in temp_vars:
    if var!= "YrSold":
        analysis_year_vars(data, var,'SalePrice')
```

```python
def analyse_discrete(df, var, target_var = "SalePrice"):
    df = df.copy()
    df.groupby(var)[target_var].median().plot.bar()
    plt.title("Medial sales price of house by {}".format(var))
    plt.ylabel("Sale Price ")
    plt.xlabel(var)
    plt.show()
```

```python
def analyze_continous(df, var):
    df = df.copy()
    df[var].hist(bins=30)
    plt.title(var)
    plt.ylabel("Number of houses")
    plt.xlabel(var)
    plt.show()

for var in cont_var:
    analyze_continous(data, var)
```

```python
def analyse_log_transformed(df, var):
    df = df.copy()
    if any(data[var] <= 0):
        pass
    else:
        df[var] = np.log(df[var])
        df[var].hist(bins=30)
        plt.ylabel("Number of houses")
        plt.xlabel(var)
        plt.show()
```

```python
for var in cont_var:
    analyse_log_transformed(data, var)
```

```python
def analyse_box_cox_transformed(df, var):
    df = df.copy()
    if any(df[var] <= 0):
        pass
    else:
        df[var], fitted_lambda = stats.boxcox(df[var])
        df[var].hist(bins=30)
        plt.ylabel("Number of houses")
        plt.xlabel(var)
        plt.show()
```

```python
for var in cont_var:
    analyse_box_cox_transformed(data, var)
```

```python
def analyse_transform_log_target(df, var, target_var):
    df = df.copy()
    if any(df[var] <= 0):
        pass
    else:
        df[var] = np.log(df[var])
        df[target_var] = np.log(df[target_var])
        plt.scatter(df[var], df[target_var])
        plt.ylabel(target_var)
        plt.xlabel(var)
        plt.title("relationship of {} to {}".format(var,target_var))
        plt.show()
```

```python
for var in cont_var:
    if var != "SalePrice":
        analyse_transform_log_target(data, var, "SalePrice")
```

# Outliers

```python
def find_outliers(df, var):
    
    if any(df[var] <= 0):
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column=var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()


for var in cont_var:
    find_outliers(data, var)
```

## Cardinality

```python
data[cat_vars].nunique()
```

```python
def analyse_rare_labels(df, var, rare_perc):
    df = df.copy()

    # determine the % of observations per category
    tmp = df.groupby(var)['SalePrice'].count() / len(df)

    # return categories that are rare
    return tmp[tmp < rare_perc]

# print categories that are present in less than
# 1 % of the observations


for var in cat_vars:
    print(analyse_rare_labels(data, var, 0.01))
    print()
```
